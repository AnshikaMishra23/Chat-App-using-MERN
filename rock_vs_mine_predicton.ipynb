Importing dependencies


```python
import numpy as np #arrays
import pandas as pd #dataframe
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Processing


```python
#loading the dataset toa pandas dataframe
sonar_data=pd.read_csv('/content/sonar_data.csv',header=None) # there was no header in the dataset
```

```python
#EDA : exploratory data analysis

import matplotlib.pyplot as plt
plt.scatter(sonar_data[0],sonar_data[1])
```

```python
#separating data and labels-- supervised learning
X=sonar_data.drop(columns=60,axis=1) #dropping a column axis =1, row axis=0
Y=sonar_data[60]
```

```python
print(X)
print(Y)
```

Training and test data split

```python
X_train, X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)

#X_train=training data
# Y_train label of X_train
#X_test=testing data
#Y_test label of X_test

#test_size=0.1 means 10 percent of data will be the test data
#stratify=Y we need to split the data based on rock and mine
#random_state=1 to specify how the data will be splitted 2 k liye doosri tarah se hoga split

```

```python
print(X.shape, X_train.shape,X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

**MODEL TRAINING** -->Logistic regression


```python
model = LogisticRegression()
# loads the LogisticRegression model to model variable
```

```python
#training the logisstic regression model with training data

model.fit(X_train,Y_train)
```

Model evaluation


```python
#accuracy check on the training data

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
#X_train_prediction- result given by model     Y_train- actual answer
#hum log dono ko compare kar rhe h

```

```python
print('Accuracy on the training data : ',training_data_accuracy)
```

```python
#accuracy check on the test data

X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on the testing data : ',testing_data_accuracy)
```
